```python
x = input("Enter a string:")
z = (str(str(x)[::-1]))
if x==z:
    print("It is a palindrome")
else:
    print("It is not a palindrome")
```

```python
x = input("Enter a string:")
result = ""
for char in x:
    if char not in result:
        result = result + char
print(result)
```
