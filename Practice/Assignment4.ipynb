```python
arr=[]
x=int(input("Enter the no. of elements:"))
for i in range(x):
    m=int(input("Enter the elements:"))
    arr.append(m)
for j in range(len(arr)-1):
    for k in range(len(arr)-j-1):
        if arr[k]>arr[k+1]:
            arr[k],arr[k+1]=arr[k+1],arr[k]
print("The sorted array is:")
print(arr)
second_largest=arr[-2]
second_smallest=arr[1]
print("Second largest element is:",second_largest)
print("Second smallest element is",second_smallest)
```

```python
def fibonacci(n):
    a=0
    b=1
    print(a,b,end="")
    #print(b)
    for i in range(2,n):
        c=a+b
        print(c,end="")
        a=b
        b=c
n = int(input("Enter the number of terms you want to print:"))
print("The fibonacci series of first",n,"terms are:")
fibonacci(n)
```

```python
def even_len(n):
    s=[]
    for i in n:
        if i%2==0:
            s.append(i)
    return s
x=input("Enter the list elements:")
n=[int(i) for i in x.split(",")]
s=even_len(n)
print(s)
```
